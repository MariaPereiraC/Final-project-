```python
def stress(stress_score):
    if stress_score <= 14:
        return 0 #normal
    elif stress_score >= 15 and stress_score <= 18:
        return 1 #mild
    elif stress_score >= 19 and stress_score <= 25:
        return 2 #moderate
    elif stress_score >= 26 and stress_score <= 33:
        return 3 #severe
    else: 
        stress_score >= 34
        return 4 #extremely severe
```

```python
data1['stress_category']= data1['stress_score'].apply(stress)
data1.head()
```

```python
def anxiety(anxiety_score):
    if anxiety_score <= 7:
        return 0 #normal
    elif anxiety_score >= 8 and anxiety_score <= 9:
        return 1 #mild
    elif anxiety_score >= 10 and anxiety_score <= 14:
        return 2 #moderate
    elif anxiety_score >= 15 and anxiety_score <= 19:
        return 3 #severe
    else: 
        anxiety_score >= 20
        return 4 #extremely severe
```

```python
data1['anxiety_category']= data1['anxiety_score'].apply(anxiety)
data1.head()
```

```python
def depression(depression_score):
    if depression_score <= 9:
        return 0 #normal
    elif depression_score >= 10 and depression_score <= 13:
        return 1 #mild
    elif depression_score >= 14 and depression_score <= 20:
        return 2 #moderate
    elif depression_score >= 21 and depression_score <= 27:
        return 3 #severe
    else: 
        depression_score >= 28
        return 4 #extremely severe
```

```python
data1['depression_category']= data1['depression_score'].apply(depression)
data1.head()
```

```python
sns.histplot(data1['stress_category'])
plt.show()
sns.histplot(data1['anxiety_category'])
plt.show()
sns.histplot(data1['depression_category'])
plt.show()
#stress is mild in the majority of the people
#anxiety and depression are extremely severe in the majority of the surveyed individuals
```
